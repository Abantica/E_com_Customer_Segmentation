```python
import datetime as dt
current_date=dt.datetime(2017,12,20)
#each cust current date - last date of transaction
#groupby will give for each cust id (unique)
RFM_scores=datasets.groupby('CustomerID').agg({'Date':lambda x :(current_date-x.max()).days,
                                            #x.max = last day of transaction
                                            'InvoiceNo': lambda x : x.count(),
                                            'Price': lambda x: x.sum()})
RFM_scores.rename(columns = {'Date' : 'Recency' , 'InvoiceNo' : 'Frequency',
                           'Price' : 'Monetory'} , inplace= True)
# inplace = True means not creating any other obj but permanent change
```

```python
'''As it is a distance algo, we need to do feature scaling. Eculidean dist is
actually pythagoras theorem. In Kmeans, we use only euclidean dist and not
others like manhattan or cosine.'''
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler_x=scaler.fit_transform(RFM_scores1)
scaler_x
```

```python
scale_data=pd.DataFrame(scaler_x , index= RFM_scores1.index , columns=RFM_scores1.columns)
#as scaled data was coming in array format and index will be 0, 1,2 so index and columns
#are given from RFMScores1
```

```python
scale_data
```

```python


from sklearn.cluster import KMeans
wcss = {}
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k, init='k-means++',max_iter=1000, random_state=101).fit(scale_data)
    scale_data["clusters"] = kmeans.labels_
    #print(data["clusters"])
    wcss[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center


    
plt.figure()
plt.plot(list(wcss.keys()), list(wcss.values()))
plt.xlabel("Number of cluster / K values")
plt.ylabel("WCSS")
plt.show()
```

```python
wcss
```

```python
from sklearn.metrics import silhouette_score

for n_cluster in range(2, 20):
    kmeans = KMeans(n_clusters=n_cluster).fit(scale_data)
    label = kmeans.labels_
    sil_coeff = silhouette_score(scale_data, label, metric='euclidean')
    print("For n_clusters={}, The Silhouette Coefficient is {}".format(n_cluster, sil_coeff))
```
